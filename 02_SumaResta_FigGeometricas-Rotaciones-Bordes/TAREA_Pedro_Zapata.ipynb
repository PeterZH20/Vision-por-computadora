3. Utilizando el archivo figuras/dog_backpack.png realizar la rotación cada 60 grados utilizando imutils.rotate e imutils.rotate_bound

    
</div> 

```python
#imutils.rotate
img = cv.imread('figuras/dog_backpack.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_rot=imutils.rotate(img, angle=60)
plt.imshow(img_rot)
plt.show()
```

```python
#imutils.rotate_bound
img = cv.imread('figuras/dog_backpack.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_rot_bound=imutils.rotate_bound(img, angle=-60)
plt.imshow(img_rot_bound)
plt.show()
```

```python
#Imagen original
img=cv.imread('figuras/monedas.jpeg')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
#Escala de grises
img=cv.imread('figuras/monedas.jpeg')
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.figure(figsize = (5,5))
plt.imshow(img_gris, cmap='gray')
plt.show()
```

```python
#Aplicar el filtro GaussianBlur
img=cv.imread('figuras/monedas.jpeg')
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_fgauss = cv.GaussianBlur(img_gris, (3,3), 0)
plt.imshow(img_fgauss, cmap='gray')
plt.show()
```

```python
#Aplicar la binarización con la opción cv.THRESH_BINARY_INV
img=cv.imread('figuras/monedas.jpeg')
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_fgauss = cv.GaussianBlur(img_gris , (3,3), 0)
thr,img_bin=cv.threshold(img_fgauss,175,255,cv.THRESH_BINARY_INV)
plt.imshow(img_bin, cmap='gray')
plt.colorbar()
plt.show()
```

```python
#Encontrar los contornos con cv.findContours(IMAGEN-BN, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE )
img=cv.imread('figuras/monedas.jpeg')
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_fgauss = cv.GaussianBlur(img_gris, (3,3), 0)
thr, img_bin = cv.threshold(img_fgauss, 
                            175,255, cv.THRESH_BINARY_INV)
contornos,_ = cv.findContours(img_bin,
                cv.RETR_EXTERNAL, 
                cv.CHAIN_APPROX_SIMPLE)
```

```python
# Visualizar los contornos  con drawContours
img=cv.imread('figuras/monedas.jpeg')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
cv.drawContours(img,contornos,-1,(255,0,0),2)
plt.imshow(img)
plt.show()
```
