## 1.5 Filtros
### 1.5.1 Filtro de la Media

```python
img = cv.imread('figuras/paisaje2.jpg')
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_fmedia = cv.blur(img_gris, (10,10))
plt.figure(figsize=(7,7))
plt.imshow(img_fmedia, cmap='gray')
plt.show()
```

### 1.5.2 Filtro de la Mediana

```python
img = cv.imread('figuras/paisaje2.jpg')
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_fmediana = cv.medianBlur(img_gris, 15)
plt.figure(figsize=(7,7))
plt.imshow(img_fmediana, cmap='gray')
plt.show()
```

### 1.5.2 Filtro de Gauss

```python
img = cv.imread('figuras/paisaje2.jpg')
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_fgauss = cv.GaussianBlur(img_gris, (3,3), 0)
plt.figure(figsize=(7,7))
plt.imshow(img_fgauss, cmap='gray')
plt.show()
```

## 1.6 Thresholding

```python
img = cv.imread('figuras/paisaje2.jpg')
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
thr,img_bin=cv.threshold(img_gris,110,255,cv.THRESH_BINARY)
plt.figure(figsize=(7,7))
plt.imshow(img_bin, cmap='gray')
plt.show()
```

## 1.7 Cambio de tamaño (escalándola a la mitad de los pixeles)

```python
img = cv.imread('figuras/paisaje2.jpg')
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
thr,img_bin=cv.threshold(img_gris,100,255,cv.THRESH_BINARY)
fraccion=0.5
columnas=int(fraccion*img_bin.shape[1])
filas=int(fraccion*img_bin.shape[0])
img_r=cv.resize(img_bin,(columnas,filas),cv.INTER_NEAREST)
plt.figure(figsize=(7,7))
plt.imshow(img_r, cmap='gray')
plt.show()
```

## 1.8 Todas las operaciones

```python
img = cv.imread('figuras/paisaje2.jpg')
#A color
img_color=cv.cvtColor(img, cv.COLOR_BGR2RGB)

#Escala de grises
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

#Filtro Media
img_fmedia = cv.blur(img_gris, (10,10))

#Filtro Mediana
img_fmediana = cv.medianBlur(img_gris, 15)

#Filtro Gauss
img_fgauss = cv.GaussianBlur(img_gris, (3,3), 0)

#Thresholding
thr,img_bin=cv.threshold(img_gris,110,255,cv.THRESH_BINARY)

#Cambiando tamaño
fraccion=0.5
columnas=int(fraccion*img_bin.shape[1])
filas=int(fraccion*img_bin.shape[0])
img_rt=cv.resize(img_bin,(columnas,filas),cv.INTER_NEAREST)
img_r=cv.resize(img_color,(columnas,filas),cv.INTER_NEAREST)

#Graficando
plt.figure(figsize=(15,10))

plt.subplot(3,3,1)
plt.imshow(img_color)
plt.title('Imagen original')

plt.subplot(3,3,2)
plt.imshow(img_gris ,cmap='gray')
plt.title('Imagen gris')

plt.subplot(3,3,3)
plt.imshow(img_bin,cmap='gray')
plt.title('Binarizada')

plt.subplot(3,3,4)
plt.imshow(img_fmedia,cmap='gray' )
plt.title('Filtro de la media')

plt.subplot(3,3,5)
plt.imshow(img_fmediana,cmap='gray' )
plt.title('Filtro de la mediana')

plt.subplot(3,3,6)
plt.imshow(img_fgauss,cmap='gray' )
plt.title('Filtro de Gauss')

plt.subplot(3,3,7)
plt.imshow(img_r,cmap='gray' )
plt.title('Cambio de tamaño (original)')

plt.subplot(3,3,8)
plt.imshow(img_rt,cmap='gray' )
plt.title('Cambio de tamaño (binarizada)')
```
