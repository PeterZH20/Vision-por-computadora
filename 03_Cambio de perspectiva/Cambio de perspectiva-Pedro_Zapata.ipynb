# 0. Librerías básicas

```python
#Librerías básicas de esta notebook
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
import pandas as pd
```

# Actividad 1
## Tomar una foto de un libro (asegurarse de que la foto tenga una perspectiva como la del ejemplo).

```python
#El nombre de la variable para guardar la imagen origina de la carta será "img"
img=cv.imread('figuras/libro.jpg')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.figure(figsize = (5,5))
plt.imshow(img)
plt.show()
```

```python
#Cambiar la imagen original a escala de grises
img=cv.imread('figuras/libro.jpg')
img_gris = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.figure(figsize = (5,5))
plt.imshow(img_gris, cmap='gray')
plt.show()
```

```python
#Cambiar la imagen original a la mitad de tamaño
img=cv.imread('figuras/libro.jpg')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
fraccion=0.5
columnas=int(fraccion*img.shape[1])
filas=int(fraccion*img.shape[0])
img_mitad=cv.resize(img,(columnas,filas),cv.INTER_NEAREST)
plt.figure(figsize=(5,5))
plt.imshow(img_mitad, cmap='gray')
plt.show()
```

```python
#Tomando la imagen original, aumentar la intensidad de la imagen (30 unidades)
img=cv.imread('figuras/libro.jpg')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
M=np.ones(img.shape,dtype='uint8')*30
img_a30=cv.add(img,M)
plt.figure(figsize=(5,5))
plt.imshow(img_a30)
plt.show()
```

```python
#Tomando la imagen original,disminuir la intensidad de la imagen (45 unidades)
img=cv.imread('figuras/libro.jpg')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
M=np.ones(img.shape,dtype='uint8')*45
img_m45=cv.subtract(img,M)
plt.figure(figsize=(5,5))
plt.imshow(img_m45)
plt.show()
```

```python
#Rotar la imagen original 180 grados
img=cv.imread('figuras/libro.jpg')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
img_rot180 = cv.flip(img, -1)
plt.figure(figsize = (5,5))
plt.imshow(img_rot180)
plt.show()
```

```python
#Rotar la imagen original con efecto espejo
img=cv.imread('figuras/libro.jpg')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
img_espejo = cv.flip(img, 0)
plt.figure(figsize = (5,5))
plt.imshow(img_espejo)
plt.show()
```

```python
#Mostrar todas las transformaciones en una sola imagen
plt.figure(figsize=(10,13))

plt.subplot(3,3,1)
plt.imshow(img)
plt.title('Imagen original')

plt.subplot(3,3,2)
plt.imshow(img_gris,cmap='gray')
plt.title('Imagen escala de grises')

plt.subplot(3,3,3)
plt.imshow(img_mitad,cmap='gray')
plt.title('Imagen escala a la mitad')

plt.subplot(3,3,4)
plt.imshow(img_a30)
plt.title('Imagen con intensidad +30')

plt.subplot(3,3,5)
plt.imshow(img_m45)
plt.title('Imagen con intensidad -45')

plt.subplot(3,3,6)
plt.imshow(img_rot180)
plt.title('Imagen rotada 180°')

plt.subplot(3,3,7)
plt.imshow(img_espejo)
plt.title('Imagen con efecto espejo')
```

# Actividad 2
Realizar el cambio de perspectiva

```python
#El nombre de la variable para guardar la imagen origina de la carta será "img"
img=cv.imread('figuras/libro.jpg')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.figure(figsize = (5,5))
plt.imshow(img)
plt.show()
```

```python
#Ampliar y mejorar la visibilidad del plano cartesiano
fig, ax = plt.subplots(figsize=(5,5))
ax.imshow(img)
ax.minorticks_on()
ax.grid(which='major',linestyle='-',linewidth='0.9',color='red')
ax.grid(which='minor',linestyle=':',linewidth='0.5',color='white')
```

```python
#Puntos de las esquinas
p3=(150,1850)
p4=(1850,2970)
p2=(2840,1100)
p1=(1550,550)

img=cv.imread('figuras/libro.jpg')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
fig, ax = plt.subplots(figsize=(5,5))
cv.circle(img,p1,35,(0,255,0),-1)
cv.circle(img,p2,35,(0,255,0),-1)
cv.circle(img,p3,35,(0,255,0),-1)
cv.circle(img,p4,35,(0,255,0),-1)
ax.imshow(img)
ax.minorticks_on()
ax.grid(which='major',linestyle='-',linewidth='0.9',color='red')
ax.grid(which='minor',linestyle=':',linewidth='0.5',color='white')
```

```python
#Obteniendo la matriz de transformación y cambiando de perspectiva
altura=img.shape[0]
ancho=img.shape[1]
img=cv.imread('figuras/libro.jpg')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)

pts1=np.float32([p1,p2,p3,p4])
pts2=np.float32([[0,0],[ancho,0],[0,altura],[ancho,altura]])

m_transformer=cv.getPerspectiveTransform(pts1,pts2)
img_nueva_perspectiva=cv.warpPerspective(img,m_transformer,(ancho,altura))
```

```python
plt.imshow(img_nueva_perspectiva)
plt.figure(figsize = (10,10))
plt.show()
```

```python
# Mostrar la imagen original y la que tiene el cambio  de perspectiva 
plt.figure(figsize=(10,13))

img=cv.imread('figuras/libro.jpg')
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Imagen original')

plt.subplot(1,2,2)
plt.imshow(img_nueva_perspectiva)
plt.title('Imagen con perspectiva')
```
